At this point, the data is still not treated, so after the pre-processing, we present this part again, but without the interference of unnecessary things, like stop words or emojis.

```python
import nltk
nltk.download('punkt', quiet=True)
from tidytext import unnest_tokens, bind_tf_idf 
from siuba import _, count, arrange
from plotnine import ggplot, aes, geom_col, geom_bar, labs, coord_flip, theme, facet_wrap
from plotnine import facet_grid, scale_y_continuous, scale_fill_manual, scale_color_manual
import warnings
from mizani.formatters import scientific_format

warnings.filterwarnings('ignore')

def tokenize(text):
    text =  nltk.WordPunctTokenizer().tokenize(text)
    return text

def discretization(value):
    if (value >= 0 and value < 1):
        value = 1
    elif (value >= 1 and value < 2):
        value = 2
    elif (value >= 2 and value < 3):
        value = 3
    elif(value >= 3 and value < 4):
        value = 4
    elif(value >= 4):
        value = 5
    return value
        
tf_idf_dataset = dataset.copy()

tf_idf_dataset["offense_rating_class"] = list(map(discretization, tf_idf_dataset.offense_rating))

polarity_dataset = tf_idf_dataset.copy()

tf_idf_dataset = (tf_idf_dataset
  >> unnest_tokens(_.word, _.text)
  >> count(_.offense_rating_class, _.word)
  >> bind_tf_idf(_.word, _.offense_rating_class, _.n)
  >> arrange(-_.tf_idf)
)


tf_idf_dataset = tf_idf_dataset.sort_values(by='tf_idf', ascending=False)

tf_idf_dataset = tf_idf_dataset.groupby(['offense_rating_class']).head(n = 10)

tf_idf_dataset['offense_rating_class'] = tf_idf_dataset['offense_rating_class'].astype('category')

(ggplot(tf_idf_dataset, aes(x = 'reorder(word, -tf_idf)', y = "tf_idf", fill = "offense_rating_class"))
    + scale_y_continuous(labels=scientific_format(digits=2))
    + geom_col()
    + labs(x = "Top Words", y = "TF-IDF Score",title = "Words characteristic of each offense rating group based on 8000 tweets")
    + coord_flip()
    + theme(legend_position = "none", panel_spacing = 1.1, figure_size=(12, 8))
    + facet_wrap("offense_rating_class", scales = "free", ncol= 3)
    + geom_bar(stat = "identity")
    + scale_fill_manual(values=["#440154", "#3b528b", "#21918b", "#5ac45e", "#dea835"])
    + scale_color_manual(values=["#440154", "#3b528b", "#21918b", "#5ac45e", "#dea835"])
)

```

## Pre-processing

As almost any dataset, this one is not perfect so we need to correct problems found during the analyzes and do the best to make it predictable and analyzable for the algorithms. 


#### Convert to lower case
In this problem context, capitalized words are not relevant, so the first step is transforming all the text content to lowercase.
To do this we use the **lower()** function.

```python
processed_data = dataset.copy()
processed_data["text"] = processed_data['text'].str.lower()
print(dataset)
print(processed_data)
```

#### Contractions, accents and emojis removed
Contractions do not represent anything relevant when compared with the full form, so using a list of know contractions we start by transforming them to the original way.

This is an English data set so any accent is unnecessary and must be removed. We do this by using the **normalize()** function that allows us to preserve the main character while removing the accent.

Finally, while reading the multiple entries, we found several emojis. These emojis represent relevant words so with the help of the **demojize()** function we convert the emoji into his title, allowing us to represent it with a word.

```python
import contractions
import unicodedata
import emoji

def contract(text):
    
    text = unicodedata.normalize('NFKD', text)
    text = "".join([c for c in text if not unicodedata.combining(c)])
    text = emoji.demojize(text)
    
    expanded_words = []    

    for word in text.split():
      # using contractions.fix to expand the shotened words
      expanded_words.append(contractions.fix(word))   
    
    expanded_text = ' '.join(expanded_words)
    return expanded_text

processed_data["text"] = list(map(contract, processed_data.text))
print(processed_data)
```

#### Removing punctuation:
For this problem any type of punctuation, code, or symbol is irrelevant. Using the **sub()** function we substitute every one of those by a space. Using space and not simply removing, prevents the fusion of words if there were no previous spaces between the words and punctuation.

```python
import re

def remove_punctuation(text):  
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)
    text = re.sub(r'\<a href', ' ', text)
    text = re.sub(r'&amp;', '', text) 
    text = re.sub(r'[_"\-;%()|+&=*%.,!?:#$@\[\]/]', ' ', text)
    text = re.sub(r'<br />', ' ', text)
    text = re.sub(r'\'', ' ', text)
    return text

processed_data["text"] = list(map(remove_punctuation, processed_data.text))

print(processed_data)
```

#### Removing stopwords
Stopwords are words that carry very little useful information for our problem. Has previously shown in the TF-IDF words plot, these words decrease the correctness of our models so they need to be removed from the text.

To decide which words must be removed, we use a list from the **stopwords** library provided by **nltk**.

```python
def remove_stopwords(text):  
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text

processed_data["text"] = list(map(remove_stopwords, processed_data.text))
print(processed_data)
```

#### Tokenizing text
To use lemmatization or stemming, all the words need to be tokenized.
This part of the code just splits each sentence into tokens (One word = One token).

```python
processed_data["text"] = list(map(tokenize, processed_data.text))

print(processed_data)
```

#### Splitting data into train and test data
We want to use cross-validation, but we also wish to compare the results with only using the **train_test_split()** function. So, in this part, we create four more datasets. Two for stemming (80% train; 20% test) and two for lemmatization (80% train; 20% test).

```python
from sklearn.model_selection import train_test_split
percentage = 0.8
data_general_lem = processed_data.copy();
data_general_stem = processed_data.copy();
data_train, data_test = train_test_split(processed_data, train_size = percentage)

data_train_lem = data_train.copy()
data_test_lem = data_test.copy()

data_train_stem = data_train.copy()
data_test_stem = data_test.copy()
```

### Lemmatization vs Stemming:
In this part we duplicate the dataset, one is treated using lemmatization and the other using stemming. In the final part, we compare the results for both.

#### Lemmatization:

```python
# Lemmatizing

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

def lemmatized_words(dataset):
    lemm = nltk.stem.WordNetLemmatizer()
    dataset['text'] = list(map(lambda word:
                                     list(map(lemm.lemmatize, word)),
                                     dataset.text))
    
lemmatized_words(data_train_lem)
lemmatized_words(data_test_lem)

def join_tokens(text):
    text = " ".join(text)
    text.strip()
    return text

data_general_lem["text"] = list(map(join_tokens, data_general_lem.text))
data_general_lem["text"] = data_general_lem["text"].str.strip()

data_train_lem["text"] = list(map(join_tokens, data_train_lem.text))
data_train_lem["text"] = data_train_lem["text"].str.strip()

data_test_lem["text"] = list(map(join_tokens, data_test_lem.text))
data_test_lem["text"] = data_test_lem["text"].str.strip()

print(data_train_lem)
print(data_test_lem)
```

#### Stemming:

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    text = ' '.join([ps.stem(w) for w in text])
    return text

data_general_stem["text"] = list(map(join_tokens, data_general_stem.text))
data_general_stem["text"] = data_general_stem["text"].str.strip()

data_train_stem["text"] = list(map(stem, data_train_stem.text))
data_train_stem["text"] = data_train_stem["text"].str.strip()

data_test_stem["text"] = list(map(stem, data_test_stem.text))
data_test_stem["text"] = data_test_stem["text"].str.strip()

print(data_train_stem)
print(data_test_stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
no_hate = data_train_stem[data_train_stem['offense_rating']< 0.1] 
lot_a_hate= data_train_stem[data_train_stem['offense_rating']>1] 

#No Hate
def get_top_unigram(corpus, n=None):
    #source: https://towardsdatascience.com/a-complete-exploratory-data-analysis-and-visualization-for-text-data-29fb1b96fb6a
    vec = CountVectorizer(ngram_range=(1, 1), stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_unigram(no_hate.text, 20)
# for word, freq in common_words:
#     print(word, freq)
df1 = pd.DataFrame(common_words, columns = ['ReviewText' , 'count'])
df1.groupby('ReviewText').sum()['count'].sort_values(ascending=False).plot(
    kind='bar', title='Top 20 bigrams in not offensive reviews after removing stop words', figsize=(14,8), color='g')
```

## Feature Engineering and Machine Learning Algorithms
For the feature engineering techniques, we decided to use a GridSearch in order to find the best ones for our models.
If the GridSearch decides for example that TF-IDF should be used, our pipeline will be set up with "use_idf=True" and the necessary changes will be made.
This pipeline does exactly the same we could do with the individual feature engineering functions so it's a way to make this document more concise.

### Using an SVR algorithm

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.svm import SVR
import time

# This is already using the "optimal" parameters
svr_lem_model = Pipeline([('vect', CountVectorizer(ngram_range = (1, 1))),
                          ('tfidf', TfidfTransformer(use_idf = False)), ('svr', SVR(kernel = "rbf", 
                            C = 1000.0, shrinking = True, gamma = "scale"))])

start_time = time.time()

svr_lem_model = svr_lem_model.fit(data_train_lem.text, data_train_lem.offense_rating)

training_svr_lem_time = (time.time() - start_time)
print("Training time for the lemmatized text: %s seconds." % training_svr_lem_time)

# =================================================================================================

# This is already using the "optimal" parameters
svr_stem_model = Pipeline([('vect', CountVectorizer(ngram_range = (1, 1))), 
                           ('tfidf', TfidfTransformer(use_idf = False)), ('svr', SVR(kernel = "rbf",
                            C = 1000.0, shrinking = True, gamma = "scale"))])

start_time = time.time()

svr_stem_model = svr_stem_model.fit(data_train_stem.text, data_train_stem.offense_rating)

training_svr_stem_time = (time.time() - start_time)
print("Training time for the stemmed text: %s seconds." % training_svr_stem_time )
```

## Neural Network

```python
import io
import os
import re
import shutil
import string
import tensorflow as tf

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
```

```python
#source: https://www.tensorflow.org/text/guide/word_embeddings


# Create a custom standardization function to strip HTML break tags '<br />'.
def custom_standardization(input_data):
  lowercase = tf.strings.lower(input_data)
  stripped_html = tf.strings.regex_replace(lowercase, '<br />', ' ')
  return tf.strings.regex_replace(stripped_html,
                                  '[%s]' % re.escape(string.punctuation), '')


# Vocabulary size and number of words in a sequence.
vocab_size = 10000
#max lenght of a sentece
sequence_length = 100

# Use the text vectorization layer to normalize, split, and map strings to
# integers. Note that the layer uses the custom standardization defined above.
# Set maximum_sequence length as all samples are not of the same length.
vectorize_layer = TextVectorization(
    standardize=custom_standardization,
    max_tokens=vocab_size,
    output_mode='int',
    output_sequence_length=sequence_length)
```

```python
embedding_dim=32

model = Sequential([
  vectorize_layer,
  Embedding(vocab_size, embedding_dim, name="embedding"),
  GlobalAveragePooling1D(),
  Dense(16, activation='relu'),
  Dense(1, activation = 'sigmoid')
])

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir="logs")
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=15)

model.compile(optimizer='adam',
              loss='mse',
              metrics=[tf.keras.metrics.RootMeanSquaredError(
    name="root_mean_squared_error", dtype=None
), 'mae'])

# Make a text-only dataset (no labels) and call adapt to build the vocabulary.
text_ds = data_train_lem.text
vectorize_layer.adapt(text_ds.to_numpy())
data_train_lem.offense_rating = data_train_lem.offense_rating/4.85

start_time = time.time()

history = model.fit(
    data_train_lem.text,
    data_train_lem.offense_rating,
    validation_split = 0.1,
    epochs=500,
    batch_size = 64,
    callbacks=[tensorboard_callback, es],
    verbose=0
)

training_nn_lem_time = (time.time() - start_time)

print("Training time for MLP model with lemmatized text: %s seconds." % training_nn_lem_time)

# ------------------------------------------------------------------------------------------------------

# Make a text-only dataset (no labels) and call adapt to build the vocabulary.
text_ds = data_train_stem.text
vectorize_layer.adapt(text_ds.to_numpy())
data_train_stem.offense_rating = data_train_stem.offense_rating/4.85

start_time = time.time()

history = model.fit(
    data_train_stem.text,
    data_train_stem.offense_rating,
    validation_split = 0.1,
    epochs=500,
    batch_size = 64,
    callbacks=[tensorboard_callback, es],
    verbose=0
)

training_nn_stem_time = (time.time() - start_time)

print("Training time for MLP model with stemmed text: %s seconds." % training_nn_stem_time)
```
